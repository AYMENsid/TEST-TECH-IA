```python
import pandas as pd
import numpy as np
```

```python
# chargement de fichier .csv 
df = pd.read_csv('all_fuels_data.csv')
# Affichage des premières lignes 
df.head()
```

```python
# verification de nombre de ligne pour chaque commodity
df['commodity'].value_counts()
```

```python
# Détection des valeur null
missing_values = df.isnull().sum()
# affichage des valeurs manquantes
missing_values
```

```python
# Détection des doublons
nb_duplicates = df.duplicated().sum()
print("Nombre de doublons détectés :", nb_duplicates)

# Suppression des doublons
df = df.drop_duplicates()
print("Taille du dataset après suppression :", df.shape)
```

```python
# Convertir la date de objet en datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Les colonnes numériques
num_cols = ['open', 'high', 'low', 'close', 'volume']
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')

# Les colonnes catégorielles
cat_cols = ['ticker', 'commodity']
df[cat_cols] = df[cat_cols].astype('object')

df.dtypes
```

```python
# Enregistrement du dataset  
df.to_csv('all_fuels_data.csv', index=False)
print("Dataset nettoyé enregistré ")
```
